```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r'C:/Users/91949/Desktop/ICT PROJECT/dataset/train.csv')
data.head()
```

## BIVARIENT ANALYSIS ON AVG ACCOUNT BALANCE V/S VINTAGE

### CHOOSE EITHER ONE

```python
plt.figure(figsize=(25, 6))
sns.lineplot(x='Vintage', y='Avg_Account_Balance', data=data)

# Set the title and labels for the plot
plt.title('Bivariate Analysis of Avg_Account_Balance and Vintage')
plt.xlabel('Vintage')
plt.ylabel('Avg_Account_Balance')

# Display the plot
plt.show()
```

### INFERENCE
#### As the vintage of the customer increases their average account balance also tends to increase. 

```python
plt.figure(figsize=(25, 6))
sns.barplot(x='Vintage', y='Avg_Account_Balance', data=data, ci = None)

# Set the title and labels for the plot
plt.title('Bivariate Analysis of Avg_Account_Balance and Vintage')
plt.xlabel('Vintage')
plt.ylabel('Avg_Account_Balance')

# Display the plot
plt.show()
```

## BIVARIENT ANALYSIS ON AVG ACCOUNT BALANCE AND REGION CODE

### CHOOSE EITHER ONE

```python
plt.figure(figsize=(25, 6))
# Create a bivariate plot of Avg_Account_Balance column with Region_Code without scatterplot and boxplot
sns.lineplot(x='Region_Code', y='Avg_Account_Balance', data=data)

# Set the title and labels for the plot
plt.title('Bivariate Analysis of Avg_Account_Balance and Region_Code')
plt.xlabel('Region_Code')
plt.ylabel('Avg_Account_Balance')

# Display the plot
plt.show()
```

```python
# Filter the dataframe to include only the 5 Region_Code

region_counts = data['Region_Code'].value_counts()
sorted_region = region_counts.sort_values(ascending=False)
top_5_region = sorted_region.head(5)
top_5_region_names = top_5_region.index.tolist()
filtered_df = data[data['Region_Code'].isin(top_5_region_names)]

# Calculate the percentage of each Region_Code
region_percentage = filtered_df['Region_Code'].value_counts(normalize=True) * 100

# Create a bivariate plot of Avg_Account_Balance column with Region_Code
sns.barplot(x='Region_Code', y='Avg_Account_Balance', data=filtered_df, ci = None)

# Set the title and labels for the plot
plt.title('Bivariate Analysis of Avg_Account_Balance and Region_Code (Top 5)')
plt.xlabel('Region_Code')
plt.ylabel('Avg_Account_Balance')

# Add percentage labels to the bars
for i, v in enumerate(region_percentage):
    plt.text(i, v, f"{v:.2f}%", ha='center', va='bottom')

# Display the plot
plt.show()
```

### Inference:
#### The customers from the top 5 region code, avg account balance percentile is shown.
