```python
# Define the age threshold to divide the data into two parts
age_threshold = 45

# Filter data for young and old customers
young_customers = data[data['Age'] <= age_threshold]
old_customers = data[data['Age'] > age_threshold]

# Combine the filtered data into a single DataFrame with an additional "Age Group" column   #CANt BE REMOVED
young_customers['Age Group'] = 'Young Customers'
old_customers['Age Group'] = 'Old Customers'
combined_data = pd.concat([young_customers, old_customers])

# Create the box plot using Seaborn
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
sns.boxplot(x='Age Group', y='Vintage', data=combined_data)

# Add labels and title
plt.xlabel('Customer Age Group')
plt.ylabel('Vintage')
plt.title('Box Plot: Age vs. Vintage for Credit Card Lead Prediction')

# Show the plot
plt.show()




```

```python
# Assuming you have the credit card data loaded into a DataFrame named 'data'

# Define the age threshold to divide the data into two parts
age_threshold = 55

# Filter data for young and old customers
young_customers = data[data['Age'] <= age_threshold]
old_customers = data[data['Age'] > age_threshold]

# Create a bar plot using Seaborn to compare the counts of young and old customers in each "Age" category for the "Vintage" variable
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
sns.barplot(x='Age', y='Vintage', data=young_customers, ci=None)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Vintage')
plt.title('Bar Plot: Age vs. Vintage for Credit Card Lead Prediction (Young Customers)')

# Show the plot
plt.show()


# Filter data for old customers with 'Vintage' above 70
old_customers_above_70 = old_customers[old_customers['Vintage'] > 70]

# Create another bar plot for old customers above 70
plt.figure(figsize=(20, 15))  # Adjust the figure size if needed
sns.barplot(x='Age', y='Vintage', data=old_customers_above_70, ci=None)

# Add labels and title
plt.xlabel('Age ')
plt.ylabel('Vintage')
plt.title('Bar Plot: Age vs. Vintage for Credit Card Lead Prediction (Old Customers above 70 Vintage)')

# Show the plot
plt.show()
```

```python
#old" customers tend to have higher median Vintage values and greater variability in their Vintage data compared to "young" customers. However, both age groups have outliers in their Vintage data, suggesting that there are individuals with longer or shorter relationships with the bank in both segments.
```

```python
#VINTAGE VS CREDIT PRODUCT

# Assuming you have the credit card data loaded into a DataFrame named 'credit_card_data'
# Convert 'Vintage' to categorical bins (adjust the bin size as needed)
data1['Vintage_Category'] = pd.cut(data1['Vintage'], bins=[0, 20, 40, 60, 80, 100, 120, 140], labels=['0-19', '20-39', '40-59', '60-79', '80-99', '100-119', '120-139'])

# Pivot the DataFrame to get the required data for the stacked bar plot
stacked_data1 = data1.pivot_table(index='Vintage_Category', columns='Credit_Product', aggfunc='size', fill_value=0)

# Calculate the percentage of each category in the 'Credit_Product' column
stacked_data1_percentage = stacked_data1.div(stacked_data1.sum(axis=1), axis=0) * 100

# Create the stacked bar plot
ax = stacked_data1_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add percentage annotations to the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, padding=3, labels=[f'{x:.1f}%' for x in container.datavalues])

# Customize the plot
plt.xlabel('Vintage Category')
plt.ylabel('% of Total')
plt.title('Vintage Category vs. Credit Product - Stacked Bar Plot')
plt.xticks(rotation=0)
plt.legend(title='Credit Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the credit card data loaded into a DataFrame named 'data'

# Divide 'Vintage' into two parts: vintage_below_70 and vintage_above_70
vintage_below_70 = data[data['Vintage'] <= 70]
vintage_above_70 = data[data['Vintage'] > 70]

# Calculate the percentage of each 'Credit_Product' category for vintage values below or equal to 70
percentage_below_70 = vintage_below_70['Credit_Product'].value_counts(normalize=True) * 100

# Calculate the percentage of each 'Credit_Product' category for vintage values above 70
percentage_above_70 = vintage_above_70['Credit_Product'].value_counts(normalize=True) * 100

# First Plot: Distribution of Credit_Product for vintage values below or equal to 70
plt.figure(figsize=(10, 6))
sns.barplot(x=percentage_below_70.index, y=percentage_below_70.values, palette='Set1')
plt.xlabel('Credit Product')
plt.ylabel('% of Total (Vintage Below or equal to 70)')
plt.title('Percentage of Credit_Product for vintage values below or equal to 70')
plt.tight_layout()

# Show the percentage values on the bars in the first plot
for idx, value in enumerate(percentage_below_70.values):
    plt.text(idx, value + 1, f'{value:.1f}%', ha='center', fontsize=10)

# Second Plot: Distribution of Credit_Product for vintage values above 70
plt.figure(figsize=(10, 6))
sns.barplot(x=percentage_above_70.index, y=percentage_above_70.values, palette='Set1')
plt.xlabel('Credit Product')
plt.ylabel('% of Total (Vintage Above 70)')
plt.title('Percentage of Credit_Product for vintage values above 70')
plt.tight_layout()

# Show the percentage values on the bars in the second plot
for idx, value in enumerate(percentage_above_70.values):
    plt.text(idx, value + 1, f'{value:.1f}%', ha='center', fontsize=10)

# Show the plots
plt.show()
```

```python
#REGION VS AVG ACCOUNT BALANCE 
# Filter the dataframe to include only the 5 Region_Code

region_counts = data['Region_Code'].value_counts()
sorted_region = region_counts.sort_values(ascending=False)
top_5_region = sorted_region.head(5)
top_5_region_names = top_5_region.index.tolist()
filtered_df = data[data['Region_Code'].isin(top_5_region_names)]

# Calculate the percentage of each Region_Code
region_percentage = filtered_df['Region_Code'].value_counts(normalize=True) * 100

# Create a bivariate plot of Avg_Account_Balance column with Region_Code
sns.barplot(x='Region_Code', y='Avg_Account_Balance', data=filtered_df, ci = None)

# Set the title and labels for the plot
plt.title('Bivariate Analysis of Avg_Account_Balance and Region_Code (Top 5)')
plt.xlabel('Region_Code')
plt.ylabel('Avg_Account_Balance')

# Add percentage labels to the bars
for i, v in enumerate(region_percentage):
    plt.text(i, v, f"{v:.2f}%", ha='center', va='bottom')

# Display the plot
plt.show()
```
